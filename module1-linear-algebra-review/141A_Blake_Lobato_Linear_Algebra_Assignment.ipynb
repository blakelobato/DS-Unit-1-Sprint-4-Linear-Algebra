## 1.1 Create a two-dimensional vector and plot it on a graph

```
import math
import matplotlib.pyplot as plt
import numpy as np
```

```
# Start with a twoD vector
blake = np.array([1, 1])

# Plot the  Vectors
plt.arrow(0, 0, blake[0], blake[1], head_width=.10, head_length=0.10, color='blue')

plt.xlim(-1.5,1.5)          
plt.ylim(-1.5,1.5)
plt.title("2D Vectors")
plt.show()
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```
# Start with a twoD vector
blake3d = np.array([[0, 0, 0, 1.5, 1.5, 1.5]])

# Plot the  Vectors
X, Y, Z, U, V, W = zip(*blake3d)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 2])
ax.set_ylim([0, 2])
ax.set_zlim([0, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
